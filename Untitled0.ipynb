```python
%matplotlib inline 

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np



url = 'https://raw.githubusercontent.com/cs109/2014_data/master/diamonds.csv'
diamonds = pd.read_csv(url, sep = ',', index_col=0)
```

```python
diamonds['price'].hist(bins=100)
```

```python
diamonds['price'].plot(kind='kde')
```

```python
colors = {'D' : 'g', 'E' : 'r', 'F' : 'b', 'G' : 'y', 'H' : 'k' , 
          'I' : 'c', 'J' : 'm'}

plt.scatter(diamonds['price'],diamonds['carat'],alpha=0.1,marker='.',
            c=[colors[x] for x in diamonds['color']])
```

```python
pd.options.display.mpl_style = 'default'
colors = {'D' : 'g', 'E' : 'r', 'F' : 'b', 'G' : 'y', 'H' : 'k' , 
          'I' : 'c', 'J' : 'm'}
plt.scatter(diamonds['price'],diamonds['carat'],alpha=0.1,marker='.',
            c=[colors[x] for x in diamonds['color']])
```

```python
diamonds['carat'].hist(bins=200)
plt.xlim(0,3)
```

```python
diamonds.plot(y='price',x='carat',
              c='c',alpha=0.1,kind='scatter',marker='.')
```

```python
diamonds.groupby('color').plot(y='price',x='carat',kind='scatter')
```

```python
diamonds.groupby('clarity').plot(y='price',x='carat',kind='scatter')
```

```python
diamonds.boxplot('price',by='clarity')
```
